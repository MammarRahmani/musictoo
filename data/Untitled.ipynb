```python
#data of top 100 Billbord songs 1965-2015

#columns = ['ID', 'WORD', 'RANKING']
top100 = pd.read_csv('lyrics.csv', encoding='latin-1')
print(top100.columns, '\n', '\n', top100.dtypes,'\n', '\n', top100.shape,'\n','\n', top100.head(5))
```

```python
songs=[]
for song in top100['Song']:
     songs.append(song)
    
print(songs[2617])
```

```python
artists=[]
for artist in top100['Artist']:
     artists.append(artist)
    
print(artists[2617])
```

```python
print(len(songs), len(artists))
```

```python
genres = []
countries = []

for i, j in zip(songs, artists):
    trackinfo = musixmatch.track_search(q_track=i, q_artist=j, page_size=10, page=1, s_track_rating='desc')
    #print(trackinfo)
    if not trackinfo['message']['body']['track_list']:
        pass
        country = ' '
        genre = ' '
        countries.append(country)
        genres.append(genre)
        print ('No data avalible')  
    else:
        trackinfo_ = trackinfo['message']['body']['track_list'][0]['track']
        artid = trackinfo_['artist_id']
        artistinfo = musixmatch.artist_get(artid)
        artistinfo_ = artistinfo['message']['body']['artist']
        country = artistinfo_['artist_country']
        countries.append(country)
        if not artistinfo_['primary_genres']['music_genre_list']:
            pass
            genre = ' '
            genres.append(genre)
            print ('No data avalible for genre')
        else: 
            genre = artistinfo_['primary_genres']['music_genre_list'][0]['music_genre']['music_genre_name']
            genres.append(genre)
    print('Music genre: ', genre, '\n', 'Artist from: ', country)
```

```python
print(len(countries), len(genres))
```

```python
songinfo = pd.DataFrame(list(zip(songs, artists, countries, genres)),
              columns=['Song','Artist', 'Country', 'Genre'])

print(songinfo.shape,'\n', songinfo.head(5))
```
